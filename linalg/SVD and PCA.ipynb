```python
def scatter2d(x, vec=None, ax=None):
    if ax is None:
        f,ax = plt.subplots(1,1, figsize=(5,5))
    plt.hold(True)
    ax.scatter(x[:,0], x[:,1])
    if vec is not None:
        ax.arrow(x.mean(axis=0)[0], x.mean(axis=0)[1], vec[0][0], vec[0][1], 
                    head_width=0.05, head_length=0.1, fc='r', ec='r')
        ax.arrow(x.mean(axis=0)[0], x.mean(axis=0)[1], vec[1][0], vec[1][1], 
                    head_width=0.05, head_length=0.1, fc='r', ec='r')
    ax.axis('equal')
```

## Interpret Matrix as drawing samples from Multivariate distribution

Before start talking SVD and PCA, it is good to discuss how to model a matrix in human mind

> Q: Let's start from $x \in \mathbb{R}^n$, what does it mean?

> A: It means x is a [ramdon vector](https://www.probabilitycourse.com/chapter6/6_1_5_random_vectors.php) sampled from some **n dimensional** multivariate distribution

```python
x = np.random.randn(1000,2)
x
```

$x$ is a matrix, how to interpret this matrix? Easy, each row is a *observation*, *example*, *instance*, each column is a *variable*, or *feature*, or *dimension*. Pick a name you want to use.

 - What is the `n` for $x \in \mathbb{R}^n$ to be true? the $n$ should be 2, which is the dimentionality of this random vector `x`
 - Let's say we draw 1000 samples from the 2d multivariate distribution, so we have 1000 observations
 
 That's all what $x$ means, let's take a look $x$:

```python
scatter2d(x)
```

In this case, the 1st dimension and 2nd dimension of $x$ are independent Guassian, we can introduce some **covariance** and **non-zero mean** into it. The word *covariance* just mean some relationship between different dimensions:

```python
x[:,1] = x[:,0] + x[:,1] + 2
```

```python
scatter2d(x)
```

Actually, this is what we usually see from unkown multivariate distribution in real life. Different dimensions are correlated!

> **The problem is**: The observer, you, just draw your samples without being aware of how different dimensions are correlated.

## SVD: Intepreting Matrix as 3 components: rotation, scaling and rotation

$Every$ matrix is $garantted$ to have a $unique$ SVD

<img  src="330px-Singular-Value-Decomposition.svg.png"/>

Now let's assume what we observed as matrix $M$, which is a transformation from identity matrix $I$. The transformation matrix `u` and `v` are [orthogonal matrix](https://en.wikipedia.org/wiki/Orthogonal_matrix). It explained the relationship between [uncorrelated and orthogonal](./Linearly independent, Orthogonal and Uncorrelated.ipynb).

```python
M = x
```

```python
u, s, v = np.linalg.svd(M)
```

```python
u.shape, s.shape, v.shape
```

$u$ and $v$ are clearly square matrix, $s$ is diagnal, further more:

```python
f,ax = plt.subplots(1,3, figsize=(15,5))
ax[0].matshow(np.dot(u,u.T))
ax[0].set_title('u is orthogonal')
ax[1].matshow(np.diag(s),extent=[0, 1, 0, 1])
ax[1].set_title('s is diagonal')
ax[2].matshow(np.dot(v,v.T), extent=[0,1,0,1])
ax[2].set_title('v is orthogonal')
```

What does S looks like?

```python
S = np.zeros((1000, 2))
S[:2,:2] = np.diag(s)
S
```

Let's reconstruct `_X` from svd and compare it to original `x`

```python
X = np.dot(np.dot(u,S),v)
```

```python
f,ax = plt.subplots(1,2, figsize=(12,4))
plt.hold(True)
ax[0].scatter(M[:,0], M[:,1])
ax[0].axis('equal')
ax[0].set_title('original')
ax[1].scatter(X[:,0], X[:,1])
ax[1].axis('equal')
ax[1].set_title('svd_reconstructed')
```

To test whether two matrices are the same, use `np.allclose`

```python
np.allclose(M, X)
```

So, X is close approximation reconstruction of M

Let's use sklearn PCA first to get a correct answer before diving into it

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2,whiten=True)
pca.fit(M)
```

```python
p = pca.components_
p
```

```python
f, ax = plt.subplots(1,2,figsize=(12,5))
scatter2d(X, p, ax[0])
scatter2d(X, v, ax[1])
```

> Now, why $p$ is different from $v$? $p$ is correct as you can see. So what is wrong with $v$?

> The answer is the principle component is really the eigen vector of covariance matrix. And covariance matrix which measures the [correlation](./Linearly independent, Orthogonal and Uncorrelated.ipynb) requires each dimension has zero-mean.

```python
v
```

```python
np.linalg.eig(np.dot(x.T,x))[1].T
```

So, $v$ is the eigen vector of non-zero-mean covariance matrix, $p$ is different:

```python
p = np.linalg.eig(np.cov(M.T))[1]
p = p.T
p
```

Do not forget, the covariance of two random variables are:  $ cov(X,Y) = E[(X-E[X])(Y-E[Y])] $

So let's do SVD again with zero-mean:

```python
_M = M - M.mean(axis=0)
_M.mean(axis=0)
```

```python
u, s, v = np.linalg.svd(_M, full_matrices=False)
```

```python
v
```

```python
p
```

Now, v and p are the same. This is how we compute `PCA` using `SVD`:

```python
f, ax = plt.subplots(1,2,figsize=(12,5))
scatter2d(M, p, ax[0])  # using sklearn PCA
scatter2d(M, v, ax[1])  # using numpy   SVD
```

```python
v.shape, M.shape
```

Now, let's do the PCA transform

```python
X = np.dot(_M, v)
```

```python
_X = pca.transform(M)
```

```python
f, ax = plt.subplots(1,2,figsize=(12,5))
scatter2d(X, cov(X.T), ax[0])
scatter2d(_X, cov(_X.T), ax[1])
```

```python
pca.components_ - v
```

So, even princeple eigen vector is the same, the transformation is still not correct!

The scaling is because of scaling matrix S, which diagnal elements are:

```python
s
```

```python
scaling = (s ** 2) / (X.shape[0]-1)
scaling
```

```python
X.var(axis=0)
```

```python
f, ax = plt.subplots(1,2,figsize=(12,5))
scatter2d(X/np.sqrt(scaling), cov(_X.T), ax[0])
scatter2d(_X, cov(_X.T), ax[1])
```

```python
np.allclose(X/np.sqrt(scaling), _X)
```

This is easy to understand mathematically

$$XV=U\Sigma$$
$$Q=XV$$
$$Q^TQ={\Sigma}^TU^TU\Sigma=\Sigma^2$$

```python
Q = np.dot(_M,v)
```

```python
# Q=US
np.allclose(Q, np.dot(u, np.diag(s)))
```

```python
np.dot(Q.T, Q)
```

```python
np.diag(s)**2
```

```python
# $Q^TQ=S^2$
np.allclose(np.dot(Q.T, Q), np.diag(s)**2)
```

Notice that $Q^TQ$ is not $np.cov(Q^T)$, function `cov` is normalized by number of samples `n-1`

```python
n = Q.shape[0]
np.cov(Q.T)*(n-1)
```

```python
np.allclose(np.cov(Q.T), np.dot(Q.T, Q)/(n-1))
```
