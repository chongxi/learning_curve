```python
from sklearn.neighbors import BallTree as BallTree
```

```python
D = np.random.randn(100000,9)
```

```python
BT = BallTree(D, leaf_size=15, metric='euclidean')
```

```python
# %%timeit
dist, ind = BT.query(D[0].reshape((1, -1)), k=15)
```

```python
dist,ind
```

```python
plot(D[ind].squeeze().T);
```
